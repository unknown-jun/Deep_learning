문제 27. AND게이트 함수를 가져와서 위에서 구한 가중치와 바이어스를 대입해서 TARGET을 잘 예측하는지 테스트 하시오


```
b = -0.35
w1 = 0.35
w2 = 0.1

def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.35, 0.1])
    b = -0.35
    tmp = np.sum(x*w) + b
    if tmp > 0:
        return 1
    else:
        return 0

print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

예제8. cost가 0이 아닐때는 가중치가 갱신될 수 있게 하시오

```
x = np.array([[-1, 0, 0], [-1, 1, 0], [-1, 0, 1], [-1, 1, 1]])
w = np.array([[0.3], [0.4], [0.1]])

target = np.array([[0], [0], [0], [1]])

def step_function(x):
    if x <= 0:
        return 0 
    else:
        return 1

def predict(x, w):
    a = np.sum(x*w.T)
    return step_function(a)

target = np.array([[0], [0], [0], [1]])

for i in range(len(x)):
    cost = target[i] - predict(x[i], w)
    if cost != 0:
        w = w + np.array([0.05 * x[i] * cost]).T
    print(w)
```

문제 29. 비용이 0이 되고 더 이상 비용이 발생하지 않았을때의 가중치가 출력되게 하시오

```
x = np.array([[-1, 0,0], [-1, 1, 0], [-1, 0, 1], [-1, 1, 1]])
w = np.array([[0.7], [0.4], [0.1]])

target = np.array([[0], [0], [0], [1]])

for j in range(5):
    sum1 = 0   # sum1에는 4개의 데이터를 입력할때 발생하는 비용(에러)가 입력될 변수
    for i in range(len(x)):
        cost = target[i] - predict(x[i], w)  # 비용을 계산하는 코드
        if cost != 0:                               # 비용이 0이 아니면
            w = w + np.array( [0.05 * x[i]*cost]).T # 가중치를 갱신
        elif cost == 0:       # 비용이 0이라면
            continue          # 그 다음 루프로 넘어감
        sum1 += cost   # sum1에 4개의 입력데이터를 사용했을 때 발생하는 비용을 입력한다.
    print(sum1, w.T)  # 보기 편하게 하려고 w.T를 함
```

문제 30. 위의 코드에서 비용(cost)가 0이 되는 시점에서 break해서 loop문을 종료하고 변경된 최종 가중치가 출력되게 하시오

```
x = np.array([[-1, 0,0], [-1, 1, 0], [-1, 0, 1], [-1, 1, 1]])
w = np.array([[0.7], [0.4], [0.1]])

target = np.array([[0], [0], [0], [1]])

for j in range(5):
    sum1 = 0   # sum1에는 4개의 데이터를 입력할때 발생하는 비용(에러)가 입력될 변수
    for i in range(len(x)):
        cost = target[i] - predict(x[i], w)  # 비용을 계산하는 코드
        if cost != 0:                               # 비용이 0이 아니면
            w = w + np.array( [0.05 * x[i]*cost]).T # 가중치를 갱신
        elif cost == 0:       # 비용이 0이라면
            continue          # 그 다음 루프로 넘어감
        sum1 += cost   # sum1에 4개의 입력데이터를 사용했을 때 발생하는 비용을 입력한다.
    if sum1==0:
        break
print(sum1, w.T)  # 보기 편하게 하려고 w.T를 함
```

문제 31. 위의 코드에 for j in range(5)를 무한 루프문으로 변경하시오

```
x = np.array([[-1, 0,0], [-1, 1, 0], [-1, 0, 1], [-1, 1, 1]])
w = np.array([[0.7], [0.4], [0.1]])

target = np.array([[0], [0], [0], [1]])

while True:
    sum1 = 0   # sum1에는 4개의 데이터를 입력할때 발생하는 비용(에러)가 입력될 변수
    for i in range(len(x)):
        cost = target[i] - predict(x[i], w)  # 비용을 계산하는 코드
        if cost != 0:                               # 비용이 0이 아니면
            w = w + np.array( [0.05 * x[i]*cost]).T # 가중치를 갱신
        elif cost == 0:       # 비용이 0이라면
            continue          # 그 다음 루프로 넘어감
        sum1 += cost   # sum1에 4개의 입력데이터를 사용했을 때 발생하는 비용을 입력한다.
    if sum1==0:
        break
        
print(sum1, w.T)  # 보기 편하게 하려고 w.T를 함
```

문제 32. 위의 코드를 아래와 같이 수행되면 수행되게끔 함수로 생성하시오 
(함수이름: perceptron_1957)  

<BR>

print( perceptron_1957(x, w, target) )  
최종 가중치: [[ 0.6, 0.5, 0.2 ]]

```
x = np.array([[-1, 0,0], [-1, 1, 0], [-1, 0, 1], [-1, 1, 1]])
w = np.array([[0.7], [0.4], [0.1]])

target = np.array([[0], [0], [0], [1]])

def perceptron_1957(x, w, target):
    while True:
        sum1 = 0   # sum1에는 4개의 데이터를 입력할때 발생하는 비용(에러)가 입력될 변수
        for i in range(len(x)):
            cost = target[i] - predict(x[i], w)  # 비용을 계산하는 코드
            if cost != 0:                               # 비용이 0이 아니면
                w = w + np.array( [0.05 * x[i]*cost]).T # 가중치를 갱신
            elif cost == 0:       # 비용이 0이라면
                continue          # 그 다음 루프로 넘어감
            sum1 += cost   # sum1에 4개의 입력데이터를 사용했을 때 발생하는 비용을 입력한다.
        if sum1==0:
            break
            
    print('가중치:',w.T)  # 보기 편하게 하려고 w.T를 함 

perceptron_1957(x, w, target) 
```
